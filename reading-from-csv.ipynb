```python
from pyspark.sql import SparkSession
from pyspark.sql import Row
from pyspark import SparkContext, SparkConf

def parseInput(lines):
    fields = lines.split(',')
    return Row(student_id=int(fields[0]), department=str(fields[1]))


if __name__ == "__main__":
    # creating spark session
    spark = SparkSession.builder.appName("Read from CSV").getOrCreate()
    
    lines = spark.sparkContext.textFile('data/input.csv')
    
    data = lines.map(parseInput)
    
    dataset = spark.createDataFrame(data)

    departmentCounts = dataset.groupBy("department").count()
    
    print(departmentCounts.collect())
```
